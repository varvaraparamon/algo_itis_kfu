    1. Выполнить обработку элементов матрицы А, имеющей N строк и М столбцов. Найти наибольшее значение среди средних значений для каждой строки матрицы. Сделать с помощью numpy (mean и max) и без него. Замерить время выполнения программы.

```python
%%timeit
a = np.random.rand(4, 5)

mean_strs = []
for i in range(4):
    x = 0
    for j in range(5):
        x += a[i][j]
    mean_strs.append(x/5)
max(mean_strs)
```

    2. Выполнить обработку элементов матрицы А, имеющей N строк и М столбцов. Найти наименьший элемент строки матрицы А, для которого сумма абсолютных значений элементов максимальна (np.argmax и np.min). Замерить время выполнения программы.

```python
%%timeit
a = np.random.rand(4, 5)
abs_sums = np.sum(np.abs(a), axis = 1)

min_i = 0
min_sum = abs_sums[0]
for i, str in enumerate(abs_sums):
    if str < min_sum:
        min_sum = str
        min_i = i

min_i
```
